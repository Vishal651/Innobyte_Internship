```python
df.drop(['New','PendingS', 'fulfilled-by','Order ID','currency','Sales Channel','ship-country'],axis=1,inplace = True)
```

```python
df.drop(['ship-country'],axis=1,inplace = True)
```

```python
df.rename(columns = {"index" : "Index", 
           "ship-service level" : "Ship-service Level",
           "ship-city" : "Ship City",
           "ship-state" : "Ship State",
           "ship-postal-code" : "Ship Postal code"}, inplace = True)
```

```python
df.isnull().sum()
```

```python
new_df = df.dropna()
```

```python
new_df.isnull().sum()
```

```python
new_df.info()
```

```python
new_df['Date'] = pd.to_datetime(new_df['Date'])
```

```python
new_df.info()
```

```python
new_df.shape
```

```python
#Exploratory Data Analysis
```

```python
new_df.describe()
```

```python
# Group by date and sum the amounts
sales_over_time = new_df.groupby('Date')['Amount'].sum()
```

```python
# Plot sales trends
plt.figure(figsize=(7,5))
sales_over_time.plot()
plt.title('Sales Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Sales Amount')
plt.grid(True)
plt.show()
```

```python
# Highest products sold in May month
```

```python
# Group by category to get the total quantity sold per category
category_analysis = new_df.groupby('Category')['Qty'].sum().reset_index()
category_analysis = category_analysis.sort_values(by='Qty', ascending=False)
```

```python
# Top 10 product categories by quantity sold
plt.figure(figsize=(7,5))
sns.barplot(data=category_analysis.head(5), x='Category', y='Qty', palette='viridis')
plt.title('Top 5 Product Categories by Quantity Sold')
plt.xlabel('Product Category')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 5 most sold product categories are T-shirt, Shirt, Blazzer, Trousers and Perfume.
```

```python
# Group by size to get the total quantity sold per size
size_analysis = new_df.groupby('Size')['Qty'].sum().reset_index()
size_analysis = size_analysis.sort_values(by='Qty', ascending=False)
```

```python
# Plot top 10 product sizes by quantity sold
plt.figure(figsize=(7, 5))
sns.barplot(data=size_analysis.head(10), x='Size', y='Qty', palette='coolwarm')
plt.title('Top 10 Product Sizes by Quantity Sold')
plt.xlabel('Product Size')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 10 product sizes are M, L, XL, XXL, S, 3XL, XS, Free, 6XL and 5xL
```

```python
# Fulfillment Analysis
fulfillment_analysis = new_df.groupby(['Fulfilment', 'Status']).size().reset_index(name='Order Count')
```

```python
# Plotting the fulfillment analysis
plt.figure(figsize=(12,7))
sns.barplot(data=fulfillment_analysis, x='Fulfilment', y='Order Count', hue='Status', palette='viridis')
plt.title('Number of Orders by Fulfillment Method and Status')
plt.xlabel('Fulfillment Method')
plt.ylabel('Order Count')
plt.legend(title='Order Status')
plt.show()
```

```python
# Most orders shipped by amazon method
```

```python
# Geographical Distribution
geo_analysis = new_df.groupby(['Ship State'])['Amount'].sum().reset_index()
geo_analysis = geo_analysis.sort_values(by='Amount', ascending=False)
```

```python
# Plotting the geographical distribution
plt.figure(figsize=(7, 5))
sns.barplot(data=geo_analysis.head(10), x='Ship State', y='Amount', palette='viridis')
plt.title('Top 10 States by Total Sales Amount')
plt.xlabel('State')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 10 states are Maharashtra, Karnataka, Telangana, Uttar pradesh, Tamilnadu, 
Delhi, Kerala, West bengal, Andhra pradsesh, Haryana by Total sales.
```
